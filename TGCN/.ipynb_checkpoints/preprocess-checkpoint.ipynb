```python
# Importing all required libraries
from nltk.corpus import stopwords
import nltk
from nltk.wsd import lesk
from nltk.corpus import wordnet as wn
import sys
import re
import pandas as pd
sys.path.append('./')
nltk.download('stopwords')
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
dataset='data'
```

```python
# User-defined functions

#cleaning string
def clean_str(st):
    st = re.sub(r"[^A-Za-z0-9(),!?\'\`]", " ", st)
    st = re.sub(r"\'s", " \'s", st)
    st = re.sub(r"\'ve", " \'ve", st)
    st = re.sub(r"n\'t", " n\'t", st)
    st = re.sub(r"\'re", " \'re", st)
    st = re.sub(r"\'d", " \'d", st)
    st = re.sub(r"\'ll", " \'ll", st)
    st = re.sub(r",", " , ", st)
    st = re.sub(r"!", " ! ", st)
    st = re.sub(r"\(", " \( ", st)
    st = re.sub(r"\)", " \) ", st)
    st = re.sub(r"\?", " \? ", st)
    st = re.sub(r"\s{2,}", " ", st)
    return st.strip().lower()

#word2vec
def loadWord2Vec(file_name):
    vocab = []
    embed = []
    word2vector_map = {}
    fp = open(file_name, 'r')
    for line in fp.readlines():
        r = line.strip().split(' ')
        if(len(r) > 2):
            vocab.append(r[0])
            vec = r[1:]
            length = len(vec)
            for i in range(length):
                vec[i] = float(vec[i])
            embed.append(vec)
            word2vector_map[r[0]] = vec
    
    print('Loaded Word Vectors!')
    fp.close()
    return vocab, embed, word2vector_map
```

```python
# Loading dataset

data = pd.read_csv('./cleaned_data2/' + dataset + '/corpus/' + dataset + '.csv')
data.head()

doc_cont_list = data.clean_tweet.to_list()
```

```python
#word frequency
word_freq = {}  

for doc_cont in doc_cont_list:
    temp = clean_str(doc_cont)
        
    words = temp.split()
    for w in words:
        if w in word_freq:
            word_freq[w] += 1
        else:
            word_freq[w] = 1
print(word_freq)
```

```python
#Cleaning doc
clean_doc = []
for doc_cont in doc_cont_list:
    temp = clean_str(doc_cont)
    words = temp.split()
    doc_words = []
    for w in words:
        if w not in stop_words and word_freq[w] >= 5:
            doc_words.append(w)
        
    doc_str = ' '.join(doc_words).strip()
    clean_doc.append(doc_str)
print(clean_doc[:1])
```

```python
#Saving corpus
clean_corpus = '\n'.join(clean_doc)
with open('./cleaned_data2/' + dataset + '/' + dataset + '_clean.txt', 'w') as f:
    f.write(clean_corpus)
```
